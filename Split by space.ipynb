To get the 'class' part into a separate column, we need to reload the original data from the Excel file first, as the 'Ship Mode' column was previously modified. Then, we can split the original 'Ship Mode' column into two new columns.

```python
# Reload the DataFrame to ensure we have the original 'Ship Mode' strings
df = pd.read_excel('superstore dataset .xlsx')

# Split the 'Ship Mode' column into two new columns using n=1 to split only at the first space
split_modes = df['Ship Mode'].str.split(' ', n=1, expand=True)

# Assign the first part to a new column, e.g., 'Ship Mode Type'
df['Ship Mode Type'] = split_modes[1]

# Display the first few rows of the DataFrame with the new columns
display(df.head())
```
