```python
import pandas as pd
df=pd.DataFrame({'name':['priya','priya2'],'Eng':[98,89],'tamil':[55,86]})
df['total']=df['Eng']+df['tamil']
df['per%']=(df['total']/3).round()
df
res=[]
for per in df['per%']:
    if per>=50:
        res.append('pass')
    elif per>=35:
        res.append('fail')
    else:
        res.append('fail')
df['result']=res
df
```
