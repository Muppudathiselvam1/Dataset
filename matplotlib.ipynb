```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np
x=['a','b','c','d']
y=[2,4,6,2]
z=[1,6,3,7]
plt.plot(x,y,color='red',linestyle='--', marker='*')
plt.plot(x,z,color='green',linestyle='--', marker='*')
plt.title('Line Plot',color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend(['jan','feb'])
plt.show()
```

```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
import numpy as np
month=['jan','feb','mar','april']
gold_price=[16000,15000,18000,20000]
colors=['green']
plt.title('gold price')
plt.xlabel('month')
plt.ylabel('price')
for i in range(len(gold_price)-1):
    if gold_price[i]>gold_price[i+1]:
      colors.append('red')
    else:
        colors.append('green')
plt.bar(month,gold_price,color=colors)
plt.show()
```
